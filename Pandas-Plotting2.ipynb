```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\singh\Downloads\alphabet_stock_data.csv")
df.head()
```

```python
start_date = pd.to_datetime('2020-04-01')
end_date = pd.to_datetime('2020-4-30')

df['Date'] = pd.to_datetime(df['Date'])
new_df = (df['Date']>=start_date) & (df['Date']<=end_date)
df1 = df.loc[new_df]
df2 = df1.set_index('Date')
```

```python
#Q3
plt.figure(figsize=(6,6))
plt.suptitle('Trading Volume of Alphabet Inc. stock,\n01-04-2020 to 30-04-2020', 
             fontsize=16, 
             color='maroon')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("Trading Volume", fontsize=12, color='black') 
df2['Volume'].plot(kind='bar');
plt.show()
```

```python
#Q4
df3 = df1[['Date', 'Open', 'Close']]
df4 = df3.set_index('Date') 

plt.figure(figsize=(20,20))
df4.plot(kind='bar');
plt.suptitle('Opening/Closing stock prices Alphabet Inc.,\n01-04-2020 to 30-04-2020', 
             fontsize=12, color='black')
plt.show()
```

```python
#Q5
plt.figure(figsize=(20,20))
df4.plot.bar(stacked=True);
plt.suptitle('Opening/Closing stock prices Alphabet Inc.,\n01-04-2020 to 30-04-2020', fontsize=12, color='black')
plt.show()
```

```python
#Q6
plt.figure(figsize=(20,20))
df4.plot.barh(stacked=True)
plt.suptitle('Opening/Closing stock prices Alphabet Inc.,\n01-04-2020 to 30-04-2020', fontsize=12, color='black')
plt.show()
```

```python
#Q7
df2 = df1[['Open','Close','High','Low']]
#df3 = df2.set_index('Date')
plt.figure(figsize=(25,25))
df2.plot.hist(alpha=0.5)
plt.suptitle('Opening/Closing/High/Low stock prices of Alphabet Inc.,\n From 01-04-2020 to 30-09-2020', 
             fontsize=12, 
             color='blue')
plt.show()
```

```python
#Q8
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-9-30') 

df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
df2 = df1[['Open','Close','High','Low']]

plt.figure(figsize=(25,25))
df2.plot.hist(stacked=True, bins=20)
plt.suptitle('Opening/Closing/High/Low stock prices of Alphabet Inc.,\n From 01-04-2020 to 30-09-2020', 
             fontsize=12, 
             color='blue')
plt.show()
```
