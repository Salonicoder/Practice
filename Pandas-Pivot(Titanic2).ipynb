```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
result = df.pivot_table(index=['sex'], columns="pclass", aggfunc='count')
result
```

```python
result = df.groupby(['sex', 'class'])['survived'].agg('mean').unstack()
result
```

```python
result = df.pivot_table('survived', index='sex', columns='class', margins=True)
result
```

```python
df.pivot_table(index='sex', columns='pclass', values='survived', aggfunc='count')
```

```python
df.pivot_table(index='sex', columns='class', aggfunc={'survived':sum, 'fare':'mean'})
```

```python
df.pivot_table('sex', 'who', aggfunc='count')
```

```python
df.loc[df['who']=='child'].isnull().sum()
```

```python
df.pivot_table( 'survived' , [ 'sex' , 'alone' ] , 'class' )
```

```python
df.pivot_table('survived', ['sex' , 'alone' ], [ 'embark_town', 'class' ])
```
