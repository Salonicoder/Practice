```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'company_code': ['Company','Company a001', '2055', 'abcd', '123345'],
    'date_of_sale ': ['12/05/2002','16/02/1999','25/09/1998','12/02/2022','15/09/1997'],
    'sale_amount': [12348.5, 233331.2, 22.5, 2566552.0, 23.0]})
df
```

```python
df['company_code_is_digit'] = list(map(lambda x: x.isdigit(), df['company_code']))
df
```

```python
del df['company_code_is_digit']
```

```python
df
```

```python
df['company_code_l_cases'] = list(map(lambda x: x.islower(), df['company_code']))
```

```python
df['company_code_u_cases'] = list(map(lambda x: x.isupper(), df['company_code']))
```

```python
df
```

```python
df.rename(columns={'company_code_ul_cases':'company_code_l_cases'}, inplace=True)
df
```

```python
del df['company_code_l_cases']
```

```python
df
```

```python
# multiple column deletion can be done using this method:
df = df.drop(columns=['company_code_u_cases'])
df
```

```python
df['company_code_is_title'] = list(map(lambda x: x.istitle(), df['company_code']))
print(df)
```

```python
df['company_code_is_space'] = list(map(lambda x: x.isspace(), df['company_code']))
print(df)
```

```python
df = df.drop(columns=['company_code_is_title', 'company_code_is_space'])
df
```
