```python
import pandas as pd
```

```python
diamonds = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.set_index('color', inplace=True)
diamonds.head()
```

```python
diamonds.index.name = 'color'
diamonds.reset_index(inplace=True)
diamonds.head()
```

```python
print("\nAccess the Series index:")
print(diamonds.carat.value_counts().index)
print("\nAccess the Series values:")
print(diamonds.carat.value_counts().values)
```

```python
print("\nSeries sorted by its values:")
print(diamonds.cut.value_counts().sort_values())
print("\nSeries sorted by its index:")
print(diamonds.cut.value_counts().sort_index())
```
