```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\singh\Downloads\alphabet_stock_data.csv")
```

```python
df.head()
```

```python
start_date = pd.to_datetime('2020-04-01')
end_date = pd.to_datetime('2020-09-30')

df['Date'] = pd.to_datetime(df['Date'])
new_df = (df['Date']>=start_date) & (df['Date']<=end_date)
```

```python
df1 = df.loc[new_df]
df2 = df1.set_index('Date')

plt.figure(figsize=(5,5))
plt.suptitle('Stock Prices of Alphabet Inc.,\n01-04-2020 to 30-09-2020',\
                fontsize=16, color='blue')
plt.xlabel("Date",fontsize=14, color='black')
plt.ylabel("$ Price", fontsize=14, color='black')

df2['Close'].plot(color='green')
plt.show()
```

```python
plt.figure(figsize=(12,14))
df1.plot(x="Date", y=['Open', 'Close'])

plt.suptitle('Opening/Closing stock prices of Alphabet Inc.,\n 01-04-2020 to 30-09-2020', 
             fontsize=14, color='blue')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
plt.show()
```
