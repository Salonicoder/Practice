```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_excel('SaleData.xlsx')
df
```

```python
pd.pivot_table(df, index=["Region", "SalesMan"])
```

```python
pd.pivot_table(df, index = ["Region", "Manager"], values = ['Sale_amt'], aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=["Region","Manager","SalesMan"], values="Sale_amt", aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=["Item"], values="Units", aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=["Region"], values="Sale_amt", aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=["Region", "Item"], values="Units", aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=["Manager"], values="Sale_amt", aggfunc=["count", "mean"])
```

```python
pd.pivot_table(df, index=["Manager", "SalesMan"], values="Sale_amt", aggfunc=np.sum, margins=True)
```

```python
pd.pivot_table(df, index=["Region", "Manager","SalesMan"], values="Sale_amt").query('Manager == ["Douglas"]')
```

```python
pd.pivot_table(df, index=["Region", "Item"], values="Units").query('Item==["Television","Home Theater"]')
```

```python
pd.pivot_table(df, index="Item", values="Sale_amt", aggfunc="max").sort_values(by=['Sale_amt'], ascending=False)
```

```python
pd.pivot_table(df, index="Item", values="Sale_amt", aggfunc="min").sort_values(by=['Sale_amt'], ascending=True)
```

```python
pd.pivot_table(df, index="Item", values="Sale_amt", aggfunc=["max","min"])
```
